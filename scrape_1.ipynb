```python
import requests
from bs4 import BeautifulSoup
import json
import time

with open("myscheme_full_500.json") as f:
    schemes = json.load(f)

base_url = "https://www.myscheme.gov.in/schemes/"

for scheme in schemes[:50]:   # test with 50 first

    slug = scheme["slug"]
    url = base_url + slug

    try:
        r = requests.get(url, headers={"User-Agent":"Mozilla/5.0"})
        soup = BeautifulSoup(r.text, "html.parser")

        text = soup.get_text(" ")

        scheme["apply_link"] = url

        # crude but works for demo
        scheme["benefits"] = text.split("Benefits")[-1][:400]
        scheme["eligibility"] = text.split("Eligibility")[-1][:400]
        scheme["documents"] = text.split("Documents Required")[-1][:400]
        scheme["application_process"] = text.split("How to Apply")[-1][:400]

        print("Done:", slug)

        time.sleep(1)

    except Exception as e:
        print("Failed:", slug, e)

with open("myscheme_enriched.json", "w") as f:
    json.dump(schemes, f, indent=2)

print("Completed")
```
